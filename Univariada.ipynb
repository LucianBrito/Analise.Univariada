```python
#criar uma base de dados contendo 5 colunas Qualitativas
import pandas as pd
import numpy as np

# Gerando dados aleatórios para o DataFrame
np.random.seed(42)  # Para resultados reproduzíveis
data = {
    "Variável 1": np.random.randint(10, 100, 10),
    "Variável 2": np.random.uniform(1.5, 10.5, 10),
    "Variável 3": np.random.normal(50, 15, 10),
    "Variável 4": np.random.poisson(5, 10),
    "Variável 5": np.random.exponential(2, 10),
}

# Criando o DataFrame
df = pd.DataFrame(data)

# Exibindo o DataFrame
print(df)
```

```python
df.head(10)
```

```python
# Calcular a média, mediana e moda dos dados

# Importando bibliotecas necessárias
from scipy import stats

# Calculando estatísticas descritivas
media = df.mean()
mediana = df.median()
moda = df.mode().iloc[0]  # O .iloc[0] retorna a primeira linha da moda

# Exibindo os resultados
print("Média:\n", media)
print("\nMediana:\n", mediana)
print("\nModa:\n", moda)
```

```python
# Calculando medidas de dispersão
variancia = df.var()
desvio_padrao = df.std()
amplitude = df.max() - df.min()
coef_variação = (desvio_padrao / df.mean()) * 100

# Exibindo os resultados
print("Variância:\n", variancia)
print("\nDesvio Padrão:\n", desvio_padrao)
print("\nAmplitude:\n", amplitude)
print("\nCoeficiente de Variação (%):\n", coef_variação)
```

```python
# Criando uma tabela resumo
resumo = pd.DataFrame({
    "Média": df.mean(),
    "Mediana": df.median(),
    "Moda": df.mode().iloc[0],
    "Variância": df.var(),
    "Desvio Padrão": df.std(),
    "Amplitude": df.max() - df.min(),
    "Coeficiente de Variação (%)": (df.std() / df.mean()) * 100,
})

# Exibindo a tabela formatada
print(resumo)

#pd.set_option('display.float_format', '{:.2f}'.format)
#print(resumo)
```

```python
# Importando bibliotecas necessárias
import pandas as pd
from scipy.stats import skew, kurtosis # Import skew and kurtosis


# Calculando o coeficiente de assimetria e curtose
assimetria = df.apply(skew)  # Coeficiente de assimetria
curtose = df.apply(kurtosis)  # Curtose

# Criando uma tabela resumo incluindo essas medidas
resumo = pd.DataFrame({
    "Média": df.mean(),
    "Mediana": df.median(),
    "Moda": df.mode().iloc[0],
    "Variância": df.var(),
    "Desvio Padrão": df.std(),
    "Amplitude": df.max() - df.min(),
    "Coeficiente de Variação (%)": (df.std() / df.mean()) * 100,
    "Assimetria": assimetria,
    "Curtose": curtose
})

# Exibindo o resumo
print(resumo)
```

```python
import seaborn as sns
import matplotlib.pyplot as plt # Import matplotlib.pyplot

for col in df.columns:
    sns.histplot(df[col], kde=True, bins=10)
    plt.title(f"Distribuição da {col}")
    plt.show()

df.boxplot(figsize=(10, 6))
plt.title("Boxplot das Variáveis")
plt.show()
```
